```python
from tkinter import *
from tkinter import ttk
def wajhaPass():
    root = Tk()
    root.title('اختر المواد المنجزة:')
    root.geometry("500x400")
    # Create A Main Frame
    main_frame = Frame(root)
    main_frame.pack(fill=BOTH, expand=1)

    # Create A Canvas
    my_canvas = Canvas(main_frame)
    my_canvas.pack(side=LEFT, fill=BOTH, expand=1)

    # Add A Scrollbar To The Canvas
    my_scrollbar = ttk.Scrollbar(main_frame, orient=VERTICAL, command=my_canvas.yview)
    my_scrollbar.pack(side=RIGHT, fill=Y)

    # Configure The Canvas
    my_canvas.configure(yscrollcommand=my_scrollbar.set)
    my_canvas.bind('<Configure>', lambda e: my_canvas.configure(scrollregion = my_canvas.bbox("all")))

    # Create ANOTHER Frame INSIDE the Canvas
    second_frame = Frame(my_canvas)

    # Add that New frame To a Window In The Canvas
    my_canvas.create_window((0,0), window=second_frame, anchor="nw")

    
    cb_vars = {}  # dict to store the BooleanVar
    for category in category_data:
        var = BooleanVar()
        l = Checkbutton(second_frame, text=category, variable=var, onvalue=True, offvalue=False)
        l.pack(anchor=CENTER)
        cb_vars[category] = var  # store the BooleanVar


    def show_selected_categories():
        global selected_pass
        selected_pass = [c for c in cb_vars if cb_vars[c].get()]
        print(selected_pass)


    Button(root, text="Check", command=show_selected_categories).pack()


    root.mainloop()
```

```python
class AcademicAdvisor(KnowledgeEngine):
    
    @Rule(NOT(Fact(StudentName=W())))
    def askName(self):
        global Global
        x=input("ما هو اسمك؟ \n")
        self.declare(Fact(StudentName=x)) 
        Global = "الاسم: " + x + "\n"
     
    @Rule(Fact(StudentName=W()))
    def askID(self):
        global Global
        global yearOfStudent
        x=input("ما هو رقمك الجامعي؟ \n")
        yearOfStudent = int(x[0:4])
        self.declare(Fact(ID =x)) 
        Global = Global +" الرقم الجامعي: " + x + "\n"
        
    @Rule(Fact(ID=W()))
    def askCourses(self):
        global Global
        wajhaPass()
        self.declare(Fact(Courses  = selected_pass)) 
        Global = Global +" المواد المنجزة: " + ' '.join(selected_pass) + "\n"
        
    @Rule(Fact(Courses=W()))
    def askHoursDone(self):
        global Global
        global Credits
        Credits = TotalHours(selected_pass)
        self.declare(Fact(HoursDone  = Credits)) 
        Global = Global +" عدد الساعات المنجزة: " + str(Credits) + "\n"
        Global = Global +"الحالة: " + str(yearLateCalc(yearOfStudent)) + "\n"
        if Credits >=100 and isJP() ==True:
            Global = Global +"بامكانك أخذ مشروع ما قبل التخرج في الفصل القادم" +"\n"
    
    @Rule (Fact(HoursDone = MATCH.a),TEST(lambda a:a >= 100))
    def askSpec(self):
        global Global
        x=input("ما هو الاختصاص؟ \n")
        self.declare(Fact(Spec=x)) 
        Global = Global +" الاختصاص: " + x + "\n"
        
    @Rule (OR(AND(Fact(HoursDone = MATCH.a),TEST(lambda a:a < 100)),Fact(Spec=W())))
    def askIfLevel(self):
        global Global
        x=input(" هل يوجد مستويات انكليزية متبقية ؟ (yes أو no)\n")
        self.declare(Fact(IfLevel=x)) 
        Global = Global +" مستويات انكليزية متبقية: " + x + "\n"
    
    @Rule (Fact(IfLevel=L("yes")))
    def askWhatLevel(self):
        global Global
        x=input("ماهو اخر مستوى قمت بانجازه ؟(0,1,2)\n")
        self.declare(Fact(WhatLevel=x)) 
        Global = Global +" اخر مستوى تم انجازه: " + str(TotalLevel(int(x))) + "\n" 
        
    @Rule (OR(Fact(IfLevel=L("no")),Fact(WhatLevel=W()))) 
    def askWhatGpa(self):
        global Global
        global GPA
        x=input("ماهو معدلك التراكمي ؟\n")
        GPA = float(x)
        self.declare(Fact(Gpa=float(x))) 
        Global = Global +" المعدل التراكمي: " + x + "\n"    
        
    @Rule (Fact(Gpa = MATCH.b),TEST(lambda b:b >= 3))
    def askDiscWarn(self):
        global Global
        x=input("هل لديك انذارات تأديبية ؟نعم(1) او لا(0)\n")
        self.declare(Fact(DiscWarn=x)) 
        Global = Global +" الانذارات التأديبية: " + x + "\n" 
        
    @Rule (Fact(Gpa = MATCH.b),TEST(lambda b:b < 2))
    def askEduStatus(self):
        global Global
        x=input(" حدد من الاختيارات التالية حالتك الدراسية ((1)تحت الانذار الأكاديمي أو (2)ضمن الدراسة الاستدراكية) \n")
        self.declare(Fact(EduStatus=x)) 
        Global = Global +" الحالة الدراسية: " + x + "\n" 
    
    @Rule (AND(Fact(Gpa = MATCH.c),TEST((lambda c:c >=2))),TEST(lambda c:c <3))
    def askConditional(self):
        global Global
        wajhaCond()
        self.declare(Fact(CondCourse=selected_cond)) 
        self.declare(Fact(CondAndBet='yes'))
        Global = Global +" المواد الشرطية: " + ' '.join(selected_cond) + "\n" 
    
    @Rule (Fact(CondAndBet=L('yes')))
    def askCondAndBetween(self):
        global Global
        Global = Global +"ننصح باعادة المواد الشرطية التالية مع فصل الفحص الوطني لرفع المعدل: " + ' '.join(selected_cond) + "\n" 
        
    @Rule (Fact(EduStatus=W()))    
    def askConditionalwithSTAT(self):
        global Global
        wajhaCond()
        self.declare(Fact(CondCourse=selected_cond)) 
        self.declare(Fact(CondAndBel='yes'))
        Global = Global +" المواد الشرطية: " + ' '.join(selected_cond) + "\n" 
    
    @Rule (AND(Fact(CondAndBel=L('yes')),Fact(EduStatus=L('1'))))  
    def askCondAndBel(self):
        x=input("  هل انت في الفصل الأول من الانذار(1) ام الفصل ثاني (2)" )
        self.declare(Fact(WarningSem=x))
        
    @Rule (Fact(WarningSem=L('1')))    
    def askCondWithOne(self):
        global Global
        Global = Global +" تبقى لك فصل واحد لرفع معدلك والا سوف تدخل في مرحلة الدراسة الاستدراكية" + "\n" 
    
    @Rule (Fact(WarningSem=L('2')))    
    def askCondWithTwo(self):
        global Global
        Global = Global +"يجب عليك رفع معدلك بفصلك الحالي والا دخلت في مرحلة الدراسة الاستدراكية" + "\n" 
        
    @Rule (AND(Fact(CondAndBel=L('yes')),Fact(EduStatus=L('2'))))  
    def askCondAndBeltwo(self):
        x=input(" هل انت في الفصل الأول من الدراسة الاستدراكية(1) ام الفصل ثاني (2)؟" )
        self.declare(Fact(WarningSemTwo=x))
     
    @Rule (Fact(WarningSemTwo=L('1')))    
    def askCondAndBeltwoWarningOne(self):
        global Global
        Global = Global +" تبقى لك فصل واحد لرفع معدلك والا سوف تنفصل" + "\n" 
    
    @Rule (Fact(WarningSemTwo=L('2')))    
    def askCondAndBeltwoWarningtwo(self):
        global Global
        Global = Global +"يجب عليك رفع معدلك بفصلك الحالي والا انفصلت" + "\n" 
        
    @Rule (Fact(CondCourse=W()))
    def askIfFailCourseWithStat(self):
        global Global
        wajhaFail()
        self.declare(Fact(FailCourse=selected_fail)) 
        Global = Global +" المواد المحمولة: " + ' '.join(selected_fail) + "\n" 
        
    @Rule (OR(Fact(FailCourse=W()),Fact(DiscWarn=W())))  
    def askNextCourse(self):
        global Global
        wajhaFuture()
        self.declare(Fact(NextCourse=selected_future))
        Global = Global +" المواد التي تود أخذها في الفصل القادم: " + ' '.join(selected_future) + "\n"
        
    @Rule (Fact(NextCourse=W()))  
    def askNextMarks(self):
        global Global
        if nextGPA(selected_future) == True:
            x=calcGPA(selected_future)
            self.declare(Fact(NextCourse=selected_future))
            Global = Global +" المعدل المتوقع: " + str(x) + "\n"
    
    @Rule (Fact(DiscWarn=L('0')))
    def askDiscWarnMnah(self):
        global Global
        if CheckCredits(Credits) == True: 
            Global = Global +" يمكنك التقديم على منح الجامعة المقدمة " +  "\n"
        
```
